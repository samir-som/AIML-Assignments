```python
#importing the neessary libraries
import numpy as np
import pandas as pd
```

```python
#reading the data in a python environment
data = pd.read_csv('/content/mobile_price_classification.csv')
```

```python
#cheking the first few rows of the data
data.head()
```

```python
#to check every column witout ....
pd.set_option('display.max_columns', None)
```

```python
data.head()
```

```python
#cheking for null values and encoding
data.info()
```

```python
x = data.drop('price_range',axis=1)
y = data['price_range']
```

```python
#so its a multi class classification problem
data['price_range'].value_counts()
```

```python
#getting train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#importing the necessary libraries
from keras.models import Sequential
from keras.layers import Dense
```

```python
#making the ann model with input layr, hidden layers and output layer
model = Sequential()
model.add(Dense(50,activation='relu',input_shape=(20,)))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='softmax'))
```

```python
#compiling the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=32,validation_data=(x_test,y_test))
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred[y_pred>0.5] = 1
y_pred[y_pred<=0.5] = 0
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
