#data preprocess

```python
#writing the function to remove the punctuations
import string
def remove_punctuation(text):
  punctuationfree = "".join([i for i in text if i not in string.punctuation])
  return punctuationfree
```

```python
#writing the function to tokenize the words
import nltk
nltk.download('punkt')
def tokenization(text):
  words = nltk.word_tokenize(text)
  return words
```

```python
#writing the function to remove the stopwords
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
  output = [i for i in text if i not in stopwords]
  return output
```

```python
#writing the function to do the lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
```

```python
#writing the function to use all the above code
def preprocess(df_col):
  corpus =[]
  for item in df_col:
    new_item = remove_punctuation(item)
    new_item = new_item.lower()
    new_item = tokenization(new_item)
    new_item = remove_stopwords(new_item)
    new_item = lemmatizer(new_item)
    corpus.append(' '.join(str(x) for x in new_item))
  return corpus
```

```python
#making the corpus
corpus = preprocess(df['tweet_text'])
```

```python
#vectorizing the corpus
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(1,2))
traindata = cv.fit_transform(corpus)
x = traindata
y = df.label
```

```python
#classification of data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
#importing the function to do the padding
from tensorflow.keras.preprocessing import sequence
```

```python
max_len = 300
```

```python
x_train = x_train.toarray()
x_test = x_test.toarray()
```

```python
x_train = np.asarray(x_train).astype(np.float32)
x_test = np.asarray(x_test).astype(np.float32)
```

```python
#doing the padding
x_train = sequence.pad_sequences(x_train,maxlen=max_len)
x_test = sequence.pad_sequences(x_test,maxlen=max_len)
```

```python
x_train.shape
```

```python
#importing the functions to create the simple rnn
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Embedding,SimpleRNN
```

#Simple RNN

```python
#creating the model
model = Sequential()
model.add(Embedding(input_dim=10000,output_dim=32,input_length=max_len))
model.add(SimpleRNN(units=32))
model.add(Dense(units=3,activation='softmax'))
model.summary()
```

```python
#compiling the model
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#fiting the model
model.fit(x_train,y_train,epochs=4,batch_size=256)
```
